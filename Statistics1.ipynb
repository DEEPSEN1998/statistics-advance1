# **Theory**

1. **What is a random variable in probability theory?**
- A random variable is a numerical outcome of a random experiment.

2. **What are the types of random variables?**
- Discrete random variables (finite/countable outcomes)
- Continuous random variables (infinite/unbounded values)

3. **What is the difference between discrete and continuous distributions?**
- Discrete: Takes specific values (e.g., number of heads in coin flips).
- Continuous: Can take any value in a range (e.g., height of people).

4. **What are probability distribution functions (PDF)?**
-  function that describes the likelihood of different values for a random variable.

5. **How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?**
- PDF: Gives the probability of a specific value (only for discrete variables).
- CDF: Gives the probability of a value being less than or equal to a given number.

6. **What is a discrete uniform distribution?**
- A distribution where each possible outcome has an equal probability (e.g., rolling a fair die).

7. **What are the key properties of a Bernoulli distribution?**
-  Two possible outcomes: success (1) and failure (0).
-    Probability of success = p, failure = 1−p.

8. **What is the binomial distribution, and how is it used in probability?**
- Models the number of successes in nn independent Bernoulli trials (e.g., flipping a coin 10 times).

9. **What is the Poisson distribution, and where is it applied?**
- Models the number of events occurring in a fixed time/space (e.g., number of emails received per hour).

10. **What is a continuous uniform distribution?**
- A distribution where all values in a range are equally likely (e.g., random numbers between 0 and 1).

11. **What are the characteristics of a normal distribution?**
- Symmetrical, bell-shaped curve
- Defined by mean μμ and standard deviation σσ
- Most values cluster around the mean

12. **What is the standard normal distribution, and why is it important?**
- A normal distribution with mean = 0 and standard deviation = 1.
- Used in Z-score calculations for standardization.

13. **What is the Central Limit Theorem (CLT), and why is it critical in statistics?**
-  The CLT states that the sampling distribution of the sample mean approaches normality as sample size increases.

14. **How does the Central Limit Theorem relate to the normal distribution?**
- Even if the population isn't normal, sample means will be approximately normally distributed.

15. **What is the application of Z statistics in hypothesis testing?**
- Z-tests are used when comparing means with known population variance.

16. **How do you calculate a Z-score, and what does it represent?**
- Z=X−μ/σ​
- Measures how many standard deviations a value is from the mean.

17. **What are point estimates and interval estimates in statistics?**

- Point estimate: A single value estimate (e.g., sample mean).
- Interval estimate: A range (e.g., confidence interval) that contains the true parameter.

18. **What is the significance of confidence intervals in statistical analysis?**
- Provides a range where the true population parameter is likely to fall.

19. **What is the relationship between a Z-score and a confidence interval?**
- Z-score helps determine confidence intervals (e.g., 95% CI uses Z = 1.96).

20. **How are Z-scores used to compare different distributions?**
- Converts different datasets into a standard scale to compare values.

21. **What are the assumptions for applying the Central Limit Theorem?**
- ndependent random samples
- Sample size is sufficiently large (n>30 is common)
- Population should not be heavily skewed for small samples

22. **What is the concept of expected value in a probability distribution?**
- The long-term average outcome of a random variable.

23. **How does a probability distribution relate to the expected outcome of a random variable?**
- The expected value is the weighted sum of all possible outcomes based on their probabilities.

```python
# 3. Bernoulli distribution PDF
def bernoulli_pdf(p, x):
    return p**x * (1 - p)**(1 - x)

p = 0.5  # Probability of success
x_values = [0, 1]
pdf_values = [bernoulli_pdf(p, x) for x in x_values]
print("Bernoulli PDF:", pdf_values)
```

```python
# 8. Normal distribution simulation
normal_data = np.random.normal(0, 1, 1000)
plt.hist(normal_data, bins=30, density=True, alpha=0.7, color='green', edgecolor='black')
plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Normal Distribution')
plt.show()
```

```python
# 9. Calculate Z-scores and plot them
def calculate_z_scores(data):
    mean = np.mean(data)
    std_dev = np.std(data)
    return (data - mean) / std_dev

z_scores = calculate_z_scores(normal_data)
plt.hist(z_scores, bins=30, density=True, alpha=0.7, color='pink', edgecolor='black')
plt.xlabel('Z-score')
plt.ylabel('Density')
plt.title('Z-score Distribution')
plt.show()
```

```python
# 10. Implement the Central Limit Theorem (CLT)
def clt_demo(sample_size, num_samples):
    means = [np.mean(np.random.exponential(scale=2, size=sample_size)) for _ in range(num_samples)]
    plt.hist(means, bins=30, density=True, alpha=0.7, color='brown', edgecolor='black')
    plt.xlabel('Sample Mean')
    plt.ylabel('Density')
    plt.title(f'Central Limit Theorem Demonstration (n={sample_size})')
    plt.show()

clt_demo(30, 1000)
```

```python
#15. Simulate multiple samples from a normal distribution and verify the Central Limit Theorem (CLT)
import numpy as np
import matplotlib.pyplot as plt

def clt_simulation(sample_size, num_samples):
    means = []
    for _ in range(num_samples):
        sample = np.random.normal(loc=0, scale=1, size=sample_size)
        means.append(np.mean(sample))

    plt.hist(means, bins=30, density=True)
    plt.title(f'CLT Simulation\nSample Size: {sample_size}, Number of Samples: {num_samples}')
    plt.show()

# Parameters: sample size = 30, number of samples = 1000
clt_simulation(30, 1000)
```

```python
#16. Calculate and plot the standard normal distribution (mean = 0, std = 1)
import scipy.stats as stats

def plot_normal_distribution():
    x = np.linspace(-5, 5, 100)
    y = stats.norm.pdf(x, 0, 1)
    plt.plot(x, y)
    plt.title('Standard Normal Distribution (mean=0, std=1)')
    plt.show()

plot_normal_distribution()
```

```python
#17. Generate random variables and calculate their corresponding probabilities using the binomial distribution
def binomial_distribution(n, p, size):
    samples = np.random.binomial(n, p, size)
    plt.hist(samples, bins=20, density=True)
    plt.title(f'Binomial Distribution\nn={n}, p={p}')
    plt.show()

# Parameters: n = 10, p = 0.5, size = 1000
binomial_distribution(10, 0.5, 1000)
```

```python
#18. Calculate the Z-score for a given data point and compare it to a standard normal distribution
def z_score(x, mean, std):
    return (x - mean) / std

# Parameters: x = 2, mean = 0, std = 1
z = z_score(2, 0, 1)
print(f'Z-score: {z}')
```

```python
#19.Implement hypothesis testing using Z-statistics for a sample dataset
from scipy import stats

def z_test(sample_mean, population_mean, std_dev, sample_size):
    z = (sample_mean - population_mean) / (std_dev / np.sqrt(sample_size))
    return z

# Parameters: sample_mean = 110, population_mean = 100, std_dev = 15, sample_size = 30
z = z_test(110, 100, 15, 30)
p_value = stats.norm.sf(abs(z))  # Two-tailed test
print(f'Z-score: {z}, P-value: {p_value}')
```

```python
#20. Create a confidence interval for a dataset using Python and interpret the result
def confidence_interval(data, confidence_level=0.95):
    mean = np.mean(data)
    std_error = stats.sem(data)
    margin_of_error = std_error * stats.t.ppf((1 + confidence_level) / 2, len(data) - 1)
    return mean - margin_of_error, mean + margin_of_error

# Generate random data from a normal distribution
data = np.random.normal(100, 15, 100)
ci = confidence_interval(data)
print(f'Confidence Interval: {ci}')
```

```python
#21. Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean
data = np.random.normal(loc=50, scale=10, size=100)
ci = confidence_interval(data)
print(f'Confidence Interval for the mean: {ci}')
```

```python
#22.  Calculate and visualize the probability density function (PDF) of a normal distribution
x = np.linspace(-5, 5, 100)
pdf = stats.norm.pdf(x, 0, 1)
plt.plot(x, pdf)
plt.title('PDF of Normal Distribution')
plt.show()
```

```python
#23.  Calculate and interpret the cumulative distribution function (CDF) of a Poisson distribution
lambda_poisson = 3  # Average rate
x = np.arange(0, 10)
cdf = stats.poisson.cdf(x, lambda_poisson)

plt.plot(x, cdf, marker='o')
plt.title('CDF of Poisson Distribution (λ = 3)')
plt.show()
```

```python
#24. Simulate a random variable using a continuous uniform distribution and calculate its expected value
def uniform_distribution(a, b, size):
    samples = np.random.uniform(a, b, size)
    expected_value = np.mean(samples)
    return expected_value

# Parameters: a = 0, b = 10, size = 1000
expected_value = uniform_distribution(0, 10, 1000)
print(f'Expected value: {expected_value}')
```

```python
#25.Compare the standard deviations of two datasets and visualize the difference
data1 = np.random.normal(0, 1, 1000)
data2 = np.random.normal(0, 2, 1000)

std_dev_1 = np.std(data1)
std_dev_2 = np.std(data2)

plt.hist(data1, bins=30, alpha=0.5, label=f'Standard Deviation: {std_dev_1}')
plt.hist(data2, bins=30, alpha=0.5, label=f'Standard Deviation: {std_dev_2}')
plt.legend()
plt.title('Comparison of Standard Deviations')
plt.show()
```

```python
#26. Calculate the range and interquartile range (IQR) of a dataset generated from a normal distribution
data = np.random.normal(0, 1, 1000)
range_data = np.max(data) - np.min(data)
iqr = stats.iqr(data)

print(f'Range: {range_data}, IQR: {iqr}')
```

```python
# 27. Implement Z-score normalization on a dataset and visualize its transformation
def z_score_normalization(data):
    return (data - np.mean(data)) / np.std(data)

data = np.random.normal(50, 15, 1000)
normalized_data = z_score_normalization(data)

plt.hist(normalized_data, bins=30, density=True)
plt.title('Z-score Normalization')
plt.show()
```

```python
#28. Calculate the skewness and kurtosis of a dataset generated from a normal distribution
from scipy.stats import skew, kurtosis

data = np.random.normal(0, 1, 1000)
skewness = skew(data)
kurt = kurtosis(data)

print(f'Skewness: {skewness}, Kurtosis: {kurt}')
```
